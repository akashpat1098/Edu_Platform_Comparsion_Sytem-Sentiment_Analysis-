```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv('WhiteHatJr_Tweets.csv')

# Count the number of tweets in each sentiment category
sentiment_counts = df['sentiment'].value_counts()


# Create a bar chart
plt.bar(sentiment_counts.index, sentiment_counts.values)

# Add chart title and labels
plt.title('WhiteHatJr_Tweets')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')

# Show the chart
plt.show()
```

```python
import pandas as pd
overall_score = []
collection_list=["WhiteHatJr_Tweets","Vedantu_Tweets","BYJUS_Tweets","Cuemath_Tweets"]

for collection in collection_list:
    # read the data from a csv file
    data = pd.read_csv('{}.csv'.format(collection))
# count the number of positive, negative, and neutral values in the sentiment column
    print(collection)
    positive_count = len(data[data['sentiment'] == 'Positive'])
    print(positive_count)
    negative_count = len(data[data['sentiment'] == 'Negative'])
    print(negative_count)
    overall_score.append(positive_count-negative_count)
print(overall_score)
m=max(overall_score)
print(m)   
```
